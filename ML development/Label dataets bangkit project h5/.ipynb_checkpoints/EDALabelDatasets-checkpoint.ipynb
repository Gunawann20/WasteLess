```python
Agg_df = df.groupby(by="kmeans_12").agg({
    'ayam-dagingT' : 'mean' ,
    'ikan-seafoodT' : 'mean' , 
    'tahu-tempe-telurT': 'mean', 
    'sayurT': 'mean',
    'sambalT': 'mean',
    'nasi-mie-pastaT': 'mean',
    'sop-soto-baksoT': 'mean',
    'kue-rotiT': 'mean',
    'jajanan-pasarT': 'mean',
    'puding-jeliT': 'mean',
    'keripik-kerupukT': 'mean',
    'buah-minumanT': 'mean'
})
```

```python
Agg_df
```

```python
Agg_df.plot.bar(figsize=(10,10))
plt.show(block=True);
```
